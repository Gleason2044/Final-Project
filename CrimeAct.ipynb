```python
import matplotlib.pyplot as plt

# Load data
monthly = pd.read_csv('Monthly Criminal Activity.csv')

# Count neighborhoods, sorted least to greatest
sort = monthly['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)

# Get the 10 lowest
bottom_10 = sort.head(10)

# Bar graph
plt.figure(figsize=(10, 6))
plt.bar(bottom_10.index, bottom_10.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Neighborhood")
plt.ylabel("Number of Incidents")
plt.title("Neighborhoods with the Fewest Criminal Incidents")
plt.tight_layout()
plt.show()
```
