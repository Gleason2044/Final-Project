The best Neighborhood would be the one that's the safest to live in, which in this dataset's case the one with the least amount of fire-related incidents 

```python
import pandas as pd

# Open data
df = pd.read_csv("FireDataSet.csv")

# Convert  to string 
df['address'] = df['address'].astype(str)

# only looks at Pittsburgh
pgh_calls = df[df['address'].str.contains('PITTSBURGH', case=False, na=False)]

# incidents
counts = pgh_calls['neighborhood'].value_counts().reset_index()
counts.columns = ['Neighborhood', 'Number of Incidents']

# fewest to most
least_incidents = counts.sort_values(by='Number of Incidents', ascending=True)

print("Neighborhoods in Pittsburgh with the fewest incidents:")

print(least_incidents.head(10))
```

Looking at this information this puts Mount Oliver Borough and Regent Square are tied as the safest neighborhoods but theres two other data sets that need to first be looked at.

The next data set to be looked at is the data that shows the crime activity in Pittsburgh

```python
import pandas as pd
import numpy as np
monthly=pd.read_csv('Monthly Criminal Activity.csv')
monthly.head(100)
sort=monthly ['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
sort.head(10)
```

This data set also confirms that My Oliver Boro is the safest neighborhood in Pittsburgh

The next data set sorts the neighborhoods based uppon the ammount of arrests they have.

```python
import pandas as pd
import numpy as np

Arrests = pd.read_csv('https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f')

Arrests.head(20)
```

```python
Arrests = Arrests.sort_values(by="INCIDENTNEIGHBORHOOD")
```

```python
Sort = Arrests['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
Sort.head(10)
```

It seems that Mt. Oliver is the neighborhood with the least ammount of arrests. This of course does not account for other outside factors like police prejudice or size of neighborhood.

With all of this data being looked at, it'd be safe to assume that Mt Oliver Borough is the safest neighborhood to live in because it has the least amount of incidents in each data set.
