The best Neighborhood would be the one that's the safest to live in, which in this dataset's case the one with the least amount of fire-related incidents 

```python
import pandas as pd

# Open data
df = pd.read_csv("FireDataSet.csv")


# only looks at Pittsburgh
pgh_calls = df[df['address'].str.contains('PITTSBURGH', case=False, na=False)]

# incidents
counts = pgh_calls['neighborhood'].value_counts().reset_index()
counts.columns = ['Neighborhood', 'Number of Incidents']

# fewest to most
least_incidents = counts.sort_values(by='Number of Incidents', ascending=True)

print("Neighborhoods in Pittsburgh with the fewest incidents:")

print(least_incidents.head(10))
```

Looking at this information this puts Mount Oliver Borough and Regent Square are tied as the safest neighborhoods but theres two other data sets that need to first be looked at.
