We chose to sort our neighborhoods based uppon the ammount of arrests.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Arrests = pd.read_csv('https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f')

Arrests.head(20)
```

```python
Arrests = Arrests.sort_values(by="INCIDENTNEIGHBORHOOD")
```

```python
Sort = Arrests['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
Sort.head(10)
```

It seems that Mt. Oliver is the neighborhood with the least ammount of arrests. This of course does not account for other outside factors like police prejudice or size of neighborhood.

```python
fig, ax = plt.subplots()
Neighborhood = ['Mt. Oliver', 'Troy Hill-Herrs Island', 'Mt. Oliver Boro', 'Central Northside']
counts = [2, 6, 18, 23]

ax.bar(Neighborhood, counts)

ax.set_ylabel('Ammount of arrests')
ax.set_title('Neighborhood')

plt.show()
```
